# creating new variable

```python
comp['isclosed'] = pd.NaT
```

```python
comp_status=[ (comp['status'] == "operating"),
              (comp['status'] == "ipo"),
              (comp['status'] =="closed"),
              (comp['status'] == "acquired") ]
#setting their respective values.
value = [1, 1, 0, 0]

#inserting values in 'isclosed' column.
comp['isclosed'] = np.select(comp_status, value)
comp                                       
```
